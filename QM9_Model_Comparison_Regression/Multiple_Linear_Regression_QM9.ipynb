#Importing all the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#Importing our dataset 'QM9_HOMO_LUMO_Gap.csv'

```python
dataset = pd.read_csv ("QM9_HOMO_LUMO_Gap.csv")
x = dataset.iloc [:, :-1].values  #This will print all the matrix of features
y = dataset.iloc [:, -1].values   #This will print all the dependent variable vectors
missing_data = dataset.isnull().sum() #This will print if there are missing_data in the dataset
```

##Printing matrix of features (x), dependent variable (y), & missing_data if available

```python
print (x)
print (y)
print (missing_data)
```

#Splitting the entire dataset into training & test sets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.2, random_state = 1 ) #Splitting the entire dataset
```

#Building our multiple linear regression model & also trained our model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression ()    #Built our model
regressor.fit (x_train, y_train)    #Trained our model using fit method
```

#Predicting the test set results by displaying results of two vectors

```python
y_pred = (regressor.predict (x_test)) #Predicted values of the x_test
np.set_printoptions (precision = 2)   #This will display our results with 2 decimal places
print (np.concatenate ((y_pred.reshape (len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1)) #This will display the results of two vectors side by side
```

#Evaluating the Model Performance (R² Score)

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```
